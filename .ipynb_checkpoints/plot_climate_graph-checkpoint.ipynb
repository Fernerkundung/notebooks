## WIP: Plot a climate graph

Example of how to plot a climate graph showing temperature as a line and precipitation as a bar graph. Data will be taken from the [WorldClim](http://www.worldclim.org/) database.

Work in progress: Need to find a good way to query the WorldClim database first.

Starting point is a [Stack Overflow answer by Joe Kington](http://stackoverflow.com/a/7353755/4169585).

```python
%matplotlib inline
import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Generate some random data
time = mdates.drange(datetime.datetime(2010, 1, 1), 
                     datetime.datetime(2011, 1, 1),
                     datetime.timedelta(days=5))
y1 = np.cumsum(np.random.random(time.size) - 0.5)
y2 = np.cumsum(np.random.random(time.size) - 0.5)
y2 = y2 * 20 + 10

# Plot things up...
fig = plt.figure()
host = fig.add_subplot(111)
par = host.twinx()

host.set_ylabel('One Thing')
par.set_ylabel('Another')

line1, = host.plot(time, y1)
line2, = par.plot(time, y2, 'go')
host.legend([line1, line2], ['Item 1', 'Item 2'])

host.xaxis_date()

# plt.show()
```
