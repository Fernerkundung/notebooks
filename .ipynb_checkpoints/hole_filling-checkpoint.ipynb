## Mathematical Morphology - Hole Filling

```python
import numpy as np
from scipy.ndimage.morphology import binary_fill_holes, binary_closing
%matplotlib inline
import matplotlib.pyplot as plt
```

## Example 1: Small Test Array

```python
hole = np.zeros((5, 5), dtype=int)
hole[1:4, 1:4] = 1
hole[2,2] = 0
hole
```

```python
filled = binary_fill_holes(hole, structure=np.ones((2,2)))
```

```python
open_hole = np.array(hole, copy=True)
open_hole[1,1] = 0
open_hole[1,2] = 0
open_hole
```

```python
open_filled = binary_fill_holes(open_hole, structure=np.ones((2,2)))
open_filled
```

```python
fig = plt.figure(figsize=(10,10))

fig.add_subplot(221)
plt.imshow(hole, interpolation="none")
plt.title("hole")

fig.add_subplot(222)
plt.imshow(filled, interpolation="none")
plt.title("filled")

fig.add_subplot(223)
plt.imshow(open_hole, interpolation="none")
plt.title("open hole")

fig.add_subplot(224)
plt.imshow(open_filled, interpolation="none")
plt.title("open hole filled")
```

## Example 2: Real Donut Image

```python
%matplotlib inline
from scipy.ndimage import imread
import matplotlib.pyplot as plt

donut_image = imread(fname="Donuts.png", flatten=True)
donut_image[donut_image < 255] = 1
donut_image[donut_image == 255] = 0
donut_image = donut_image.astype(np.int8)

plt.imshow(donut_image)
```

```python
from scipy.ndimage.morphology import binary_closing

donut_closed = binary_closing(donut_image, structure=np.ones((3,2)))
plt.imshow(donut_closed)
```

```python
from scipy.ndimage.morphology import binary_closing

donut_filled = binary_fill_holes(donut_closed)
plt.imshow(donut_filled)
```
